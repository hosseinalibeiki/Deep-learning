```python
import matplotlib.pyplot as plt
import tensorflow as tf
```

# New Section

```python
def relu(z):
    return tf.maximum(0, z)

def plot_relu(x, z, ylabel):
    plt.plot(x, relu(z))
    plt.title('ReLU Activation Function')
    plt.xlabel('x')
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()

x = tf.linspace(-3, 3, 1000)
z = 2 * x + 3
plot_relu(x, z, ylabel='ReLU(2x + 3)')
```
