```python
import matplotlib.pyplot as plt
import tensorflow as tf
```

# Relu function

```python
def relu(z):
    return tf.maximum(0, z)

def plot_relu(x, z, ylabel):
    plt.plot(x, relu(z))
    plt.title('ReLU Activation Function')
    plt.xlabel('x')
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()

x = tf.linspace(-3, 3, 1000)
z = 2 * x + 3
plot_relu(x, z, ylabel='ReLU(2x + 3)')
```

```python
def relu(z):
    return tf.maximum(0, z) #The expression tf.maximum(0, z) is commonly used in TensorFlow to apply a ReLU (Rectified Linear Unit) activation function to a tensor z.
```

```python
def plot_relu(x, z, ylabel, subplot_position):
    plt.subplot(2, 3, subplot_position)
    plt.plot(x, relu(z))
    plt.xlabel('x')
    plt.ylabel(ylabel)
    plt.grid(True)
```

```python
x = tf.linspace(-3,3,1000)
```

```python
# Define three hidden layers
z1 = 2*x+8
z2 = -4*x+2
z3 = -5*x+6
```

```python
z1r=relu(z1)
```

```python
z1 == z1r
```

```python
# Define weights for each hidden layer
phi0 = 0.5
phi1 = -1.2
phi2 = -0.8
phi3 = 0.6
```

```python
# Sum the outputs of the three hidden layers with their respective weights
z_sum = phi0 + phi1 * relu(z1) + phi2 * relu(z2) + phi3 * relu(z3)
```

```python
plt.figure(figsize=(10, 7))
```

```python
plot_relu(x, z1, r'ReLU(2x + 8)', 1)
plot_relu(x, z2, r'ReLU(-4x + 2)', 2)
plot_relu(x, z3, r'ReLU(-5x + 6)', 3)
```

```python
plt.subplot(2,1,2)
plt.plot(x,z_sum)
plt.title(r'$0.5 - 1.2 h_1 - 0.8 h_2 + 0.6 h_3$')
plt.xlabel('x')
plt.ylabel('output,y')
plt.grid(True)
plt.tight_layout()
plt.show()
```
